### Line Plot

```python
x = np.array([3, 11, 16, 23, 27, 34, 48, 55, 56, 67])
y = np.array([13, 23, 34, 49, 47, 61, 69, 78, 84, 89])
```

```python
z = np.array([87,77,75,71,54,33,21,23,19,10])
```

```python
plt.figure(figsize = (16,9))
plt.title("my first plot")
plt.xlabel("first value")
plt.ylabel("second value")
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.xticks(range(0,81,10))
plt.yticks(range(0,101,10))
plt.plot(x,y, marker = "o", color = "magenta" , ls = ":", lw = 2, mfc = "Green")
```

```python
plt.figure(figsize = (16,9))
plt.title("my first plot")
plt.xlabel("first value")
plt.ylabel("second value")
plt.grid()
plt.plot(x, y, label = 2022, color = "blue")
plt.plot(x, z, label = 2023, color = "green")
plt.legend()
```

```python
plt.figure(figsize = (16,9))
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,z)
```
