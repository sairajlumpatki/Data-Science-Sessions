```python
df = pd.read_csv("mtcars.csv")
```

### seperate the input and output

```python
x = df[["disp", "hp", "wt"]]
y = df["mpg"]
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

### create the object

```python
reg = LinearRegression()
```

### train the model

```python
reg.fit(x,y)
```

### predict on unseen data

```python
disp = 221
hp = 501
wt = 2.51
reg.predict([[disp, hp, wt]])
```

```python
reg.score(x,y)
```

## Decision Tree Classification

### import the dataset

```python
df = pd.read_csv("Social_Network_Ads.csv")
```

```python
df.shape
```

```python
df
```

### seperate input and output variables

```python
x = df[["Age", "EstimatedSalary"]]
y = df["Purchased"]
```

```python
x;
```

```python
y.value_counts()
```

### visualize the data

```python
plt.title("purchased data")
plt.xlabel("Age")
plt.ylabel("salary")
plt.grid()
plt.scatter(x["Age"], x["EstimatedSalary"], c = y)
```

### import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

###  create the object

```python
classifier = DecisionTreeClassifier(random_state = 0)
```

### train the model

```python
classifier.fit(x,y)
```

### plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12,15))
plot_tree(classifier, fontsize = 7, feature_names = ["age", 'sal'], class_names = ["y", "n"], filled = True, rounded =True);
```

```python
new = pd.DataFrame({
    "Age": [23,45,54,67], "EstimatedSalary" : [99999,88888,70000,150000]
})
```

```python
new
```

```python
classifier.predict(new)
```
