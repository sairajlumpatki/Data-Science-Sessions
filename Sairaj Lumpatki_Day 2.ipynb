### mean

```python
import numpy as np
num = np.array([60,40,20,10,70,50,30,90,10])
np.mean(num)
```

### median

```python
np.sort(num)
np.median(num)
```

## variance 

```python
np.mean(num-num.mean()**2)
```

### standerd deviation

```python
num = np.array([60,40,20,10,70,50,30,90,10])
np.sqrt(np.mean((num-num.mean())**2))
```

```python
num.std()
```

```python
np.std(num)
```
